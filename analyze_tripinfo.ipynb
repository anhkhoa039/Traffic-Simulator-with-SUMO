```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('tripinfo.csv', sep=';')

# Print summary statistics
print("Average travel time:", df['tripinfo_duration'].mean())
print("Max travel time:", df['tripinfo_duration'].max())
print("Min travel time:", df['tripinfo_duration'].min())

# Plot travel time per vehicle
plt.figure(figsize=(10, 6))
plt.bar(df['tripinfo_id'], df['tripinfo_duration'], color='skyblue')
plt.axhline(y=df['tripinfo_duration'].mean(), color='red', linestyle='--', label='Avg Travel Time')
plt.xlabel('Vehicle ID')
plt.ylabel('Travel Time (seconds)')
plt.title('Travel Time per Vehicle')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV (assuming it's comma-separated, adjust if needed)
df = pd.read_csv("tripinfo.csv", sep=";")

# Preview the data
print("Tripinfo Summary:")
print(df.describe())

# ------------------------------
# Plot 1: Trip Duration vs Vehicle ID
plt.figure(figsize=(10, 6))
plt.bar(df['tripinfo_id'], df['tripinfo_duration'], color='skyblue')
plt.xlabel('Vehicle ID')
plt.ylabel('Trip Duration (s)')
plt.title('Trip Duration per Vehicle')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("trip_duration_per_vehicle.png")
plt.show()

# ------------------------------
# Plot 2: Time Loss vs Vehicle ID
plt.figure(figsize=(10, 6))
plt.bar(df['tripinfo_id'], df['tripinfo_timeLoss'], color='salmon')
plt.xlabel('Vehicle ID')
plt.ylabel('Time Loss (s)')
plt.title('Time Loss per Vehicle')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("time_loss_per_vehicle.png")
plt.show()

# ------------------------------
# Plot 3: Waiting Time vs Vehicle ID
plt.figure(figsize=(10, 6))
plt.bar(df['tripinfo_id'], df['tripinfo_waitingTime'], color='orange')
plt.xlabel('Vehicle ID')
plt.ylabel('Waiting Time (s)')
plt.title('Waiting Time per Vehicle')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("waiting_time_per_vehicle.png")
plt.show()

# ------------------------------
# Plot 4: Average Speed per Vehicle (routeLength / duration)
df['avg_speed'] = df['tripinfo_routeLength'] / df['tripinfo_duration']
plt.figure(figsize=(10, 6))
plt.bar(df['tripinfo_id'], df['avg_speed'], color='green')
plt.xlabel('Vehicle ID')
plt.ylabel('Average Speed (m/s)')
plt.title('Average Speed per Vehicle')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("average_speed_per_vehicle.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

tripinfo_static_path = "output/static_trafficlight_output/tripinfo.csv"
tripinfo_adaptive_path = "output/adaptive_trafficlight_output/tripinfo.csv"
static_df = pd.read_csv(tripinfo_static_path, sep=';')
adaptive_df = pd.read_csv(tripinfo_adaptive_path, sep=';')

# Select comparable vehicles only (matching trip IDs)
common_ids = set(static_df['tripinfo_id']) & set(adaptive_df['tripinfo_id'])
static_df = static_df[static_df['tripinfo_id'].isin(common_ids)].sort_values(by='tripinfo_id')
adaptive_df = adaptive_df[adaptive_df['tripinfo_id'].isin(common_ids)].sort_values(by='tripinfo_id')

# Compare metrics
def compare_metric(metric):
    plt.figure(figsize=(12, 6))
    plt.plot(static_df['tripinfo_id'], static_df[metric], label='Static', marker='o')
    plt.plot(adaptive_df['tripinfo_id'], adaptive_df[metric], label='Adaptive', marker='x')
    plt.xticks(rotation=90)
    plt.xlabel('Vehicle ID')
    plt.ylabel(metric)
    plt.title(f"Comparison of {metric} per vehicle (Static vs Adaptive)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{metric}_comparison.png")
    plt.show()

# Metrics to compare
metrics = ['tripinfo_duration', 'tripinfo_timeLoss', 'tripinfo_waitingTime']
for metric in metrics:
    compare_metric(metric)

# Summary stats
summary = pd.DataFrame({
    'Metric': metrics,
    'Static Mean': [static_df[m].mean() for m in metrics],
    'Adaptive Mean': [adaptive_df[m].mean() for m in metrics],
    'Static Max': [static_df[m].max() for m in metrics],
    'Adaptive Max': [adaptive_df[m].max() for m in metrics],
})

print("\nðŸ“Š Summary Comparison:\n")
print(summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

tripinfo_static_path = "output/static_trafficlight_output/tripinfo.csv"
tripinfo_adaptive_path = "output/adaptive_trafficlight_output/tripinfo.csv"
static_df = pd.read_csv(tripinfo_static_path, sep=';')
adaptive_df = pd.read_csv(tripinfo_adaptive_path, sep=';')

# Match common trip IDs
common_ids = sorted(set(static_df['tripinfo_id']) & set(adaptive_df['tripinfo_id']))
static_df = static_df[static_df['tripinfo_id'].isin(common_ids)].sort_values(by='tripinfo_id')
adaptive_df = adaptive_df[adaptive_df['tripinfo_id'].isin(common_ids)].sort_values(by='tripinfo_id')

# Set bar width
bar_width = 0.35
x = np.arange(len(common_ids))  # x-ticks

# Plot each metric
def plot_bar_comparison(metric, ylabel, title):
    static_vals = static_df[metric].values
    adaptive_vals = adaptive_df[metric].values

    plt.figure(figsize=(14, 6))
    plt.bar(x - bar_width/2, static_vals, width=bar_width, label='Static', color='tomato')
    plt.bar(x + bar_width/2, adaptive_vals, width=bar_width, label='Adaptive', color='mediumseagreen')

    plt.xlabel('Vehicle ID')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(x, common_ids, rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.grid(axis='y')
    plt.savefig(f"{metric}_bar_comparison.png")
    plt.show()

# Plot for each metric
plot_bar_comparison('tripinfo_duration', 'Duration (s)', 'Trip Duration Comparison: Static vs Adaptive')
plot_bar_comparison('tripinfo_timeLoss', 'Time Loss (s)', 'Time Loss Comparison: Static vs Adaptive')
plot_bar_comparison('tripinfo_waitingTime', 'Waiting Time (s)', 'Waiting Time Comparison: Static vs Adaptive')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load tripinfo CSVs
tripinfo_static_path = "output/static_trafficlight_output/tripinfo.csv"
tripinfo_adaptive_path = "output/adaptive_trafficlight_output/tripinfo.csv"
tripinfo_adaptive_path = "tripinfo.csv"

static_df = pd.read_csv(tripinfo_static_path, sep=';')
adaptive_df = pd.read_csv(tripinfo_adaptive_path, sep=';')

# Match common trip IDs
common_ids = sorted(set(static_df['tripinfo_id']) & set(adaptive_df['tripinfo_id']))
static_df = static_df[static_df['tripinfo_id'].isin(common_ids)].sort_values(by='tripinfo_id')
adaptive_df = adaptive_df[adaptive_df['tripinfo_id'].isin(common_ids)].sort_values(by='tripinfo_id')

# Setup for plotting
bar_width = 0.35
x = np.arange(len(common_ids))
metrics = ['tripinfo_duration', 'tripinfo_timeLoss', 'tripinfo_waitingTime']
titles = ['Trip Duration (s)', 'Time Loss (s)', 'Waiting Time (s)']

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True)

for i, metric in enumerate(metrics):
    ax = axes[i]
    static_vals = static_df[metric].values
    adaptive_vals = adaptive_df[metric].values

    ax.bar(x - bar_width/2, static_vals, width=bar_width, label='Static', color='salmon')
    ax.bar(x + bar_width/2, adaptive_vals, width=bar_width, label='Adaptive', color='mediumseagreen')

    ax.set_title(titles[i])
    ax.set_xlabel('Vehicle ID')
    ax.set_ylabel('Value')
    ax.grid(axis='y')

    ax.set_xticks(x)
    ax.set_xticklabels(common_ids, rotation=90)

# Add legend once
axes[0].legend(loc='upper left')

plt.suptitle('ðŸš¦ Static vs Adaptive Traffic Light Performance per Vehicle', fontsize=16)
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.savefig("static_vs_adaptive_summary.png")
plt.show()
```
